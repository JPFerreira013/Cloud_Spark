```python
import findspark
findspark.init()
import pyspark
sc = pyspark.SparkContext()
```

```python
file = "/content/quijote.txt"
```

```python
lineas = sc.textFile(file)
```

```python
lineas.take(5)
```

```python
lineas.count()
```

```python
listas_palavras = lineas.map(lambda elemento : elemento.split())
```

```python
listas_palavras.take(5)
```

```python
palavras_linea = lineas.map(lambda elemento:len(elemento))
palavras_linea.take(5)
```

```python
#Palavras por linea
palabras_linea = listas_palavras.map(lambda elemento: len(elemento))

palabras_linea.take(5)
```

```python
palavras_linea.reduce(lambda e1, e2: e1 + e2)
```

```python
#Flat map no devolve os colchetes de separaçao

lista_palabras = lineas.flatMap(lambda elemento: elemento.split())
lista_palabras.take(17)
```
